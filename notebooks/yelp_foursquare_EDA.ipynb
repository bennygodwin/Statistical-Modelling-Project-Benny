# Foursquare

Send a request to Foursquare with a small radius (1000m) for all the bike stations in your city of choice. 

```python
headers = {
    "accept": "application/json",
    # remove the api key and commit
    "Authorization": ""
}

Yelpheaders = {
    "accept": "application/json",
    # remove the api key and commit
    "Authorization": ""
}
```

```python
data = []
```

```python
count = 0
for idx, item in df.iterrows():
    url = "https://api.foursquare.com/v3/places/search?ll=" + \
        item["latitude"].__str__()+"%2C"+item["longitude"].__str__() + \
        "&radius=1000"

    response = requests.get(url, headers=headers).json()
    for x in response['results']:
        temp = {
            "fsq_id": x['fsq_id'],
            "name": x['name'],
            "latitude":item["latitude"],
            "longitude":item["longitude"],
            "categories": ""
        }
        i = 0
        for y in x['categories']:
            temp["categories"] = temp["categories"] + y["name"] + ","
            i += 1

        if temp not in data:
            data.append(temp)
            
    print(count)
    count += 1
```

Parse through the response to get the POI (such as restaurants, bars, etc) details you want (ratings, name, location, etc)

```python
len(data)
```

```python
i = 0
for x in data:

    url = "https://api.foursquare.com/v3/places/" + \
        x['fsq_id']+"?fields=rating%2Cprice%2Cpopularity%2Cstats"
    tempRes = requests.get(url, headers=headers).json()
    try:
        data[i]["popularity"] = tempRes['popularity']
    except:
        pass
    try:
        data[i]["rating"] = tempRes['rating']
    except:
        pass
    try:
        data[i]["stats"] = tempRes['stats']
    except:
        pass
    try:
        data[i]["price"] = tempRes['price']
    except:
        pass
    i += 1

    print(i)

```

```python
data.__len__()
```

Put your parsed results into a DataFrame

```python
df = pd.DataFrame.from_dict(data)
```

```python
df.head()
```

```python
df.to_csv("../data/fourSquare.csv",header=['fsqid','name','latitude','longitude','category','popularity','price','rating','stats'])
```
