Build a regression model.

```python
import pandas as pd
import statsmodels.api as sm
import numpy as np
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv("../data/finalDf.csv")
df.drop("Unnamed: 0", axis=1, inplace=True)
df.head()
```

```python
df.fillna(0,inplace=True)
```

```python
df.head()
```

```python
labelencoder = LabelEncoder()
df['category_new'] = labelencoder.fit_transform(df['category'].astype(str))
df.head(5)
```

```python
# Prepare the inputs for the regression model
x = df[['latitude', 'longitude','popularity','price','rating','category_new']]
x = sm.add_constant(x) # Add a constant term to the inputs
y = df['free_bikes']
```

```python
# Fit the regression model
model = sm.OLS(y, x).fit()
```

Provide model output and an interpretation of the results. 

```python
# Print the summary
print(model.summary())
```
